```python
import pandas as pd
import numpy as np 
```

```python
df=pd.DataFrame(
    [[2.4,np.nan],[6.3,-5.4],
     [np.nan,np.nan],[0.75,-1.3]],
    index=["a","b","c","d"],
    columns=["one","two"])
df
```

```python
df.sum() #sum of each column
```

```python
df.sum(axis=1) #sum of each row
```

```python
df.mean(axis=1) #mean of each row
```

```python
df.mean(axis=1,skipna=False)   #mean of each row, but skipna=False, so that NaN values are not ignored later,between a number and Nan is Nan
```

```python
df.idxmax()        #index of the maximum value in each column
```

```python
df.idxmin()       #index of the minimum value in each column
```

```python
df.cumsum()     #cumulative sum of each column (2.4+6.30=8.7+0.75=9.45) (-5.4-1.3=-6.7)
```

```python
df.describe()   #summary statistics of each column
```

To find the correlation coefficient, let's first import the famous iris data set. You can download iris data set from https://archive.ics.uci.edu/ml/datasets/iris.

```python
iris=pd.read_csv("DataSets/iris.data", 
                 sep=",",
                 header=None)
```

```python
iris.head()
```

```python
iris.columns=['sepal_length','sepal_width',
             'petal_length','petal_width',
             'class']
```

```python
iris.head()
```

```python
iris["sepal_length"].corr(iris["sepal_width"])
```

```python

numeric_columns = iris.select_dtypes(include=[float, int])

correlation_matrix = numeric_columns.corr()

print(correlation_matrix)
# iris.corr() Con el string de class no me deja hacer la correlaci√≥n
```

```python
# Seleccionar solo las columnas num√©ricas
numeric_columns = iris.select_dtypes(include=[float, int])

# Calcular la covarianza entre las columnas num√©ricas
covariance_matrix = numeric_columns.cov()

print(covariance_matrix)
# iris.cov() #covariance matrix
```

```python
# Seleccionar solo las columnas num√©ricas
numeric_columns = iris.select_dtypes(include=[float, int])

# Calcular la correlaci√≥n con la columna 'petal_length'
correlation_with_petal_length = numeric_columns.corrwith(iris['petal_length'])

print(correlation_with_petal_length)


# iris.corrwith(iris.petal_length) #correlation of all columns with petal_length
```
