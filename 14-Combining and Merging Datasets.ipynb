```python
import pandas as pd
import numpy as np
```

## Joining DataFrame

```python
d1=pd.DataFrame(
    {"key":["a","b","c","c","d","e"],
     "num1":range(6)})
d2=pd.DataFrame(
    {"key":["b","c","e","f"],
     "num2":range(4)})
```

```python
print(d1)
print(d2)
```

```python
pd.merge(d1, d2)  #coge las letras que tienen en com√∫n y ponen los n√∫meros de ambas tablas
```

```python
pd.merge(d1, d2, on='key') #coge las letras que tienen en com√∫n y ponen los n√∫meros de ambas tablas
```

```python
d3=pd.DataFrame(
    {"key1":["a","b","c","c","d","e"],
     "num1":range(6)})
d4=pd.DataFrame(
    {"key2":["b","c","e","f"],
     "num2":range(4)})
```

```python
pd.merge(
    d3,d4,left_on="key1",right_on="key2"
)
```

```python
print(d1)
print(d2)
pd.merge(d1,d2,how="outer") # merge con todas las letras de ambas tablas    
```

```python
pd.merge(d1,d2,how="left") # merge con todas las letras de la tabla de d1
```

```python
pd.merge(d1,d2,how="right") # merge con todas las letras de la tabla de d2
```

```python
pd.merge(d1, d2, how='inner') # merge con las letras que tienen en com√∫n
```

```python
df1=pd.DataFrame(
    {"key":["a","b","c","c","d","e"],
     "num1":range(6),
     "count":["one","three","two",
              "one","one","two"]})
df2=pd.DataFrame(
    {"key":["b","c","e","f"],
     "num2":range(4),
     "count":["one","two","two","two"]}) 
print (df1)
print (df2)
```

```python
pd.merge(df1, df2, on=['key', 'count'],  
         how='outer')  # merge con todas las letras de ambas tablas, tanto si tiene mismo count o no 
```

```python
pd.merge(df1, df2, on="key", how='outer') #ahora anexiona en una fila aunque no tengan mismp count, si tienen msima letra
```

```python
pd.merge(df1, df2, 
         on='key', 
         suffixes=('_data1', '_data2'))   #a√±ade sufijos a las columnas que se repiten
```

## Merging on index

```python
df1=pd.DataFrame(
    {"letter":["a","a","b",
               "b","a","c"],
     "num":range(6)}) 
df2=pd.DataFrame(
    {"value":[3,5,7]},
    index=["a","b","e"])
```

```python
print(df1)
print(df2)
```

```python
pd.merge(df1,df2,
         left_on="letter",
         right_index=True)  #merge con las letras que tienen en com√∫n y ponen los n√∫meros de ambas tablas
```

```python
right=pd.DataFrame(
    [[1,2],[3,4],[5,6]],
    index=["a","c","d"],
    columns=["Tom","Tim"])
left=pd.DataFrame(
    [[7,8],[9,10],[11,12],[13,14]],
    index=["a","b","e","f"],
    columns=["Sam","Kim"]) 

print(right)    
print(left)
```

```python
pd.merge(right,left, 
         right_index=True, 
         left_index=True, 
         how="outer") #merge con todas las letras de ambas tablas
```

```python
left.join(right) #pone primero las columnas de la tabla de la izquierda (int)
```

```python
left.join(right,how="outer") #pone primero las columnas de la tabla de la izquierda (float)
```

```python
data=pd.DataFrame([[1,3],[5,7],[9,11]],            
                  index=["a","b","f"],      
                  columns=["Alex","Keta"])
left.join([right,data]) #a metido data a la derecha
```

## Concatenating Along an Axis

```python
seq= np.arange(20).reshape((4, 5)) 

print(seq)
```

```python
np.concatenate([seq,seq], axis=1) #concatena las tablas ("a la izquierda")
```

```python
np.concatenate([seq, seq], axis=0) #concatena las tablas("debajo")
```

```python
data1 = pd.Series(
    [0, 1], index=['a', 'b'])
data2 = pd.Series(
    [2,3,4], index=['c','d','e'])
data3 = pd.Series(
    [5, 6], index=['f', 'g'])
```

```python
pd.concat([data1,data2,data3]) #concatena las tablas("debajo")
```

```python
pd.concat([data1, data2, data3], axis=1) #concatena las tablas en funci√≥n de los √≠ndices (0=data1, 1=data2, 2=data3)
```

```python
data4= pd.Series([10,11,12], 
                 index=['a','b',"c"])
pd.concat([data1,data4],axis=1,join="inner")
```

```python
x=pd.concat([data1, data2, data4], 
            keys=['one', 'two','three']) #concatena las tablas y les a√±ade una clave
x
```

```python
x=pd.concat([data1, data2, data4], 
            axis=1,
            keys=['one', 'two', 'three']) #concatena las tablas y les a√±ade una clave(como fila)
x
```

```python
df1 = pd.DataFrame(
    np.arange(6).reshape(3, 2),
    index=['a', 'b', 'c'],
    columns=['one', 'two'])
df2 = pd.DataFrame(
    10+np.arange(4).reshape(2,2),
    index=['a', 'c'],
    columns=['three', 'four'])

print (df1)

print (df2)
```

```python
pd.concat([df1, df2], axis=1, 
          keys=['s1', 's2'],
          sort=False) #concatena las tablas y les a√±ade una clave
```

```python
data1 = pd.DataFrame(
    np.random.randn(3, 4),
    columns=['a','b','c','d'])
data2 = pd.DataFrame(
    np.random.randn(2, 3),
    columns=['b','d','a'])

print(data1)
print(data2)
```

```python
pd.concat([data1, data2], ignore_index=True) #las columnas en com√∫n las pone en el mismo orden
```
