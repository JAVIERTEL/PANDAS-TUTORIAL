## What is DataFrame?

```python
import pandas as pd
```

```python
data={"name":["Bill","Tom","Tim","John","Alex","Vanessa","Kate"],      
      "score":[90,80,85,75,95,60,65],      
      "sport":["Wrestling","Football","Skiing","Swimming","Tennis",
               "Karete","Surfing"],      
      "sex":["M","M","M","M","F","F","F"]}
```

```python
df=pd.DataFrame(data) #Dataframe is a container for Series
```

```python
df
```

```python
df=pd.DataFrame(data,columns=["name","sport","sex","score"])  #Change the order of columns
df
```

```python
df.head() #Show the first 5 rows
```

```python
df.tail()  #Show the last 5 rows
```

```python
df.tail(3) #Show the last 3 rows
```

```python
df.head(2) #Show the first 2 rows
```

```python
df=pd.DataFrame(data,columns=["name", "sport", "gender", "score", "age"]) #Add a new column
df
```

```python
df=pd.DataFrame(data,columns=["name", "sport", "gender", "score", "age"],
                index=["one","two","three","four","five","six","seven"]) #Add an index
df
```

```python
df["sport"] #Select a column
```

```python
my_columns=["name","sport"]
df[my_columns] #Select multiple columns
```

```python
df.sport #Select a column
```

```python
df.loc[["one"]] #Select a row
```

```python
df.loc[["one","two"]] #Select multiple rows
```

```python
df["age"]=18 #Assign a value to a column
```

```python
df=pd.DataFrame(data,columns=["name", "sport", "gender", "score", "age"], 
                index=["one","two","three","four","five","six","seven"])
values=[18,19,20,18,17,17,18]
df["age"]=values
df #Assign a list to a column
```

```python
df["pass"]=df.score>=70 #Assign a boolean value to a column
df
```

```python
del df["pass"] #Delete a column
df
```

```python
scores={"Math":{"A":85,"B":90,"C":95}, "Physics":{"A":90,"B":80,"C":75}} #Nested dictionary 
```

```python
scores_df=pd.DataFrame(scores) #Create a dataframe from a nested dictionary
scores_df
```

```python
scores_df.T #Transpose a dataframe
```

```python
scores_df.index.name="name" #Add a name to the index
scores_df.columns.name="lesson" #Add a name to the columns
```

```python
scores_df
```

```python
scores_df.values #Get the values of a dataframe
```

```python
scores_index=scores_df.index #Get the index of a dataframe

print(scores_index)
```

```python
new_index=["Jack","Tom","Bill"]

new_index[1]="Jack"
scores_df.index = new_index
print("Nuevo √≠ndice:", scores_df.index)
print(scores_df)
```
